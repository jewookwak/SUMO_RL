```python
import pandas as pd

file_reward ='C:/Users/jewoo/OneDrive/바탕 화면/highway_episodic/DQN_LSTM/data_log/1/log_data_reward.csv'
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn
```

```python
reward = pd.read_csv(file_reward,index_col=0)
reward.set_index('time',inplace=True)
reward['R1'] = reward['r1'].cumsum()
reward['R2'] = reward['r2'].cumsum()
reward['R2_1'] = reward['r2_1'].cumsum()
reward['R3'] = reward['r3'].cumsum()
reward['R4'] = reward['r4'].cumsum()
reward['R5'] = reward['r5'].cumsum()
reward['R5_1'] = reward['r5_1'].cumsum()
reward['R7'] = reward['r7'].cumsum()
reward['R8'] = reward['r8'].cumsum()
reward['Total'] = reward['total'].cumsum()
```

```python
reward['R1'].plot(figsize =(8,2),title='Collision negative reward',color = 'red')
plt.show()
reward['R2'].plot(figsize =(8,2),title='LC_SD negative reward',color = 'red')
plt.show()
reward['R2_1'].plot(figsize =(8,2),title='Forward_SD negative reward',color = 'red')
plt.show()
reward['R3'].plot(figsize =(8,2),title='Toward biggest empty space positive reward')
plt.show()
reward['R4'].plot(figsize =(8,2),title='Rear vehicle approaching negative reward',color = 'red')
plt.show()
reward['R5'].plot(figsize =(8,2),title='Mandatory LC complete positive reward')
plt.show()
reward['R5_1'].plot(figsize =(8,2),title='Unnecessary LC complete negative reward',color = 'red')
plt.show()
reward['R7'].plot(figsize =(8,2),title='Target Lane approaching positive reward')
plt.show()
reward['R8'].plot(figsize =(8,2),title='No collision rear vehicle passing positive reward')
plt.show()
reward['Total'].plot(figsize =(8,2),title='Total reward',color = 'purple')
plt.show()
```
