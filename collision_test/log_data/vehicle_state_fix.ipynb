```python
import pandas as pd
file = 'C:/Users/jewoo/OneDrive/바탕 화면/SUMO/highway/log_data/log_data.csv'
```

```python
v_state = pd.read_csv(file,index_col=0)
v_state.set_index('time',inplace=True)
```

```python
accel0=[]
accel1=[]
truck0=[]
truck1=[]
car0=[]
car1=[]
car2=[]
car3=[]
```

```python
for i in range(12):
    accel0.append(v_state[v_state['vehicle_id']=='accel0.'+str(i)]) 
    truck0.append(v_state[v_state['vehicle_id']=='truck0.'+str(i)]) 
for i in range(8):
    accel1.append(v_state[v_state['vehicle_id']=='accel1.'+str(i)]) 
    truck1.append(v_state[v_state['vehicle_id']=='truck1.'+str(i)]) 
for i in range(15):
    car0.append(v_state[v_state['vehicle_id']=='car0.'+str(i)]) 
    car1.append(v_state[v_state['vehicle_id']=='car1.'+str(i)]) 
    car2.append(v_state[v_state['vehicle_id']=='car2.'+str(i)]) 
    car3.append(v_state[v_state['vehicle_id']=='car3.'+str(i)]) 
```

```python
for i in range(12):
    accel0[i]['a_x'].iloc[0] =0
    accel0[i]['a_y'].iloc[0] =0
    for j in range(accel0[i].shape[0]-1):
        accel0[i]['a_x'].iloc[j+1] =(accel0[i]['v_x'].iloc[j+1]-accel0[i]['v_x'].iloc[j])
        accel0[i]['a_y'].iloc[j+1] =(accel0[i]['v_y'].iloc[j+1]-accel0[i]['v_y'].iloc[j])
        
for i in range(12):
    truck0[i]['a_x'].iloc[0] =0
    accel0[i]['a_y'].iloc[0] =0
    for j in range(truck0[i].shape[0]-1):
        truck0[i]['a_x'].iloc[j+1] =(truck0[i]['v_x'].iloc[j+1]-truck0[i]['v_x'].iloc[j])
        truck0[i]['a_y'].iloc[j+1] =(truck0[i]['v_y'].iloc[j+1]-truck0[i]['v_y'].iloc[j])
        
for i in range(8):
    accel1[i]['a_x'].iloc[0] =0
    accel1[i]['a_y'].iloc[0] =0
    for j in range(accel1[i].shape[0]-1):
        accel1[i]['a_x'].iloc[j+1] =(accel1[i]['v_x'].iloc[j+1]-accel1[i]['v_x'].iloc[j])
        accel1[i]['a_y'].iloc[j+1] =(accel1[i]['v_y'].iloc[j+1]-accel1[i]['v_y'].iloc[j])
        
for i in range(8):
    truck1[i]['a_x'].iloc[0] =0
    truck1[i]['a_y'].iloc[0] =0
    for j in range(truck1[i].shape[0]-1):
        truck1[i]['a_x'].iloc[j+1] =(truck1[i]['v_x'].iloc[j+1]-truck1[i]['v_x'].iloc[j])
        truck1[i]['a_y'].iloc[j+1] =(truck1[i]['v_y'].iloc[j+1]-truck1[i]['v_y'].iloc[j])
        
for i in range(15):
    car0[i]['a_x'].iloc[0] =0
    car0[i]['a_y'].iloc[0] =0
    for j in range(car0[i].shape[0]-1):
        car0[i]['a_x'].iloc[j+1] =(car0[i]['v_x'].iloc[j+1]-car0[i]['v_x'].iloc[j])
        car0[i]['a_y'].iloc[j+1] =(car0[i]['v_y'].iloc[j+1]-car0[i]['v_y'].iloc[j])
for i in range(15):
    car1[i]['a_x'].iloc[0] =0
    car1[i]['a_y'].iloc[0] =0
    for j in range(car1[i].shape[0]-1):
        car1[i]['a_x'].iloc[j+1] =(car1[i]['v_x'].iloc[j+1]-car1[i]['v_x'].iloc[j])
        car1[i]['a_y'].iloc[j+1] =(car1[i]['v_y'].iloc[j+1]-car1[i]['v_y'].iloc[j])
for i in range(15):
    car2[i]['a_x'].iloc[0] =0
    car2[i]['a_y'].iloc[0] =0
    for j in range(car2[i].shape[0]-1):
        car2[i]['a_x'].iloc[j+1] =(car2[i]['v_x'].iloc[j+1]-car2[i]['v_x'].iloc[j])
        car2[i]['a_y'].iloc[j+1] =(car2[i]['v_y'].iloc[j+1]-car2[i]['v_y'].iloc[j])
for i in range(15):
    car3[i]['a_x'].iloc[0] =0
    car3[i]['a_y'].iloc[0] =0
    for j in range(car3[i].shape[0]-1):
        car3[i]['a_x'].iloc[j+1] =(car3[i]['v_x'].iloc[j+1]-car3[i]['v_x'].iloc[j])
        car3[i]['a_y'].iloc[j+1] =(car3[i]['v_y'].iloc[j+1]-car3[i]['v_y'].iloc[j])
```
